GEFILTERTE TABELLE

```python
columns = ['OBJECTID', 'UIDENTSTLAE', 'ULAND', 'UREGBEZ', 'UMONAT', 'USTUNDE', 
           'UWOCHENTAG', 'UKATEGORIE', 'UART', 'UTYP1', 'ULICHTVERH', 
           'IstStrassenzustand', 'IstRad', 'XGCSWGS84', 'YGCSWGS84']

df_filtered = df[(df['ULAND'] == 11) & (df['IstRad'] == 1)]

df_selected = df_filtered[columns]

df_filtered 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Liste der numerischen Spalten, die du plotten möchtest
columns_to_plot = ['UREGBEZ', 'UMONAT', 'USTUNDE', 'UWOCHENTAG', 'ULICHTVERH', 'IstStrassenzustand', 'UKATEGORIE', 'UART']

# Berechnung der Anzahl der Subplots
num_plots = len(columns_to_plot)
num_rows = int(np.ceil(num_plots / 2))  # Berechne die Anzahl der Zeilen
num_cols = 2  # Zwei Spalten

# Erstellen der Subplots
fig, ax = plt.subplots(num_rows, num_cols, figsize=(16, num_rows * 4))

# Flatten der Axes-Matrix, um einfacher darauf zugreifen zu können
ax = ax.flatten()

# Durchlaufen der Spalten und Plotten der Histogramme
for i, val in enumerate(columns_to_plot):
    # Histogramm für die jeweilige Spalte plotten
    df_selected.hist(column=val, ax=ax[i], bins=20, grid=False)
    
    # Titel für jedes Plot setzen
    ax[i].set_title(f'Histogram of {val}')

# Entfernen der leeren Subplots (falls vorhanden)
for j in range(num_plots, len(ax)):
    fig.delaxes(ax[j])

# Layout anpassen, um Überlappungen zu vermeiden
plt.tight_layout()
plt.show()
```
