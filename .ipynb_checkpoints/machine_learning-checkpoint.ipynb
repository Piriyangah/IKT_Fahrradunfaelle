```python
real_df = pd.read_csv("data/accidents_2022/Unfallorte2022_LinRef.csv", sep=";")
```

```python
near_df = pd.read_csv("data/Berlin-incidents.csv", on_bad_lines="skip")
```

```python
near_df.head()
```

```python
real_df.head()
```

```python
near_accidents = near_accidents[["lon", "lat", "near_accidents"]].copy()
```

```python
near_accidents
```

```python
real_df = real_df[(real_df["ULAND"] == 11) & (real_df["IstRad"] == 1)].reset_index().copy()
```

```python
real_df = real_df.rename(columns={"XGCSWGS84": "lon", "YGCSWGS84": "lat"})
```

```python
real_df["lon"] = real_df["lon"].str.replace(",",".")
real_df["lat"] = real_df["lat"].str.replace(",",".")

real_df["lon"] = real_df["lon"].astype(float)
real_df["lat"] = real_df["lat"].astype(float)
```

```python
merged_df = real_df.merge(near_accidents, on= ["lon", "lat"], how="left")
```
